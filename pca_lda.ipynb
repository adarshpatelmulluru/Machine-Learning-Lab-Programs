```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
dataframe = load_iris()
x = dataframe.data
y=dataframe.target
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
correlationmatrix = np.corrcoef(x_scaled.T)
plt.figure(figsize=(10,6))
ax = sns.heatmap(correlationmatrix,annot=True)
ax.set_xticklabels(['sepal_length','sepal_width','petal_length','petal_width'])
ax.set_yticklabels(['sepal_length','sepal_width','petal_length','petal_width'])
plt.show()
```

```python
pca = PCA(n_components=2)
x_projected = pca.fit_transform(x_scaled)

print(f" shape of original data : {x.shape}")
print(f"shape of data after pda : {x_projected.shape}")
```

```python
pca_1 = x_projected[:,0]
pca_2 = x_projected[:,1]

plt.scatter(pca_1, pca_2, c=y, cmap="jet")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title('pca1 vs pca2')
```

```python
lda = LDA(n_components=2)
x_projected_lda = lda.fit_transform(x_scaled,y)

print(f"shape before lda of data {x_scaled.shape}")
print(f"shape after lda on data {x_projected_lda.shape}")
```

```python
lda_1 = x_projected_lda[:,0]
lda_2 = x_projected_lda[:,1]

plt.figure(figsize=(10,6))
plt.scatter(lda_1,lda_2,c=y,cmap="jet")
plt.xlabel('linear discriminant 1')
plt.ylabel('linear discriminant 2')
plt.title('lda1 vs lda2')
plt.show()
```
